## Train/Test Split Expts

```python
def train_test_split_indices(_num_expts, _train_test_split_ratio):
    all_expts = list(range(_num_expts))
    random.shuffle(all_expts)
    
    # Calculate the number of training experiments
    num_train_expts = int(_num_expts * _train_test_split_ratio)
    
    # Split the list into train and test indices
    train_indices = all_expts[:num_train_expts]
    test_indices = all_expts[num_train_expts:]
    
    return sorted(train_indices), sorted(test_indices)
train_indices,test_indices= train_test_split_indices(_num_expts=num_total_expts, _train_test_split_ratio=train_test_split_ratio)
print("Training indices:", train_indices)
print("Test indices:", test_indices)
```

```python
all_expt_obs_frames = [] # observed output torques & velocities
all_expt_cmd_vel_frames = [] # commanded velocities
for  expt_idx, bag_path in enumerate(bag_path_set):
    b = bagreader(bag_path_set[expt_idx])
    for t in b.topics:
        if 'joint_state' in t:
            joint_state_df = pd.read_csv(b.message_by_topic(t))
        elif 'joint_velocity' in t:
            joint_cmd_vel_df = pd.read_csv(b.message_by_topic(t))
        else:
            raise ValueError("Unhandled topic in ROS Bag")
    
    assert joint_state_df is not None and joint_cmd_vel_df is not None

    joint_state_df.sort_values(by=['Time'], inplace = True)
    joint_cmd_vel_df.sort_values(by=['Time'], inplace = True)

    joint_state_df = cc.standardize_time(joint_state_df)
    joint_cmd_vel_df = cc.standardize_time(joint_cmd_vel_df)
    
    joint_state_df['expt_idx'] = expt_idx
    joint_cmd_vel_df['expt_idx'] = expt_idx

    all_expt_obs_frames.append(joint_state_df.copy(deep=True))
    all_expt_cmd_vel_frames.append(joint_cmd_vel_df.copy(deep=True))


assert len(all_expt_obs_frames) == len(all_expt_cmd_vel_frames)
```

## Annotate

```python
# list of list of lists indexed by the expt_idx
all_collision_ranges = [] 
for expt_idx in range(coll_timestamps_df.shape[0]):

    filt_coll_timestamps = coll_timestamps_df[coll_timestamps_df['expt_idx'] == expt_idx]

    assert filt_coll_timestamps.shape[0] == 1

    start_time = filt_coll_timestamps.loc[filt_coll_timestamps.index[0], 'start_time']
    end_time = filt_coll_timestamps.loc[filt_coll_timestamps.index[0], 'end_time']

    start_time_dt = datetime.strptime(start_time, '%Y-%m-%d %H:%M:%S')
    end_time_dt = datetime.strptime(end_time, '%Y-%m-%d %H:%M:%S')
    
    # Assuming the timezone is Australia/Sydney
    australia_tz = pytz.timezone('Australia/Sydney')
    
    start_time_dt = australia_tz.localize(start_time_dt)
    end_time_dt = australia_tz.localize(end_time_dt)

    # add collision column
    joint_torques_df = all_expt_obs_frames[expt_idx]
    joint_torques_df['collisions'] =joint_torques_df['time'].between(start_time_dt.timestamp(), end_time_dt.timestamp()).astype(int)

    if expt_idx > num_inter_expts:
         assert joint_torques_df['collisions'].sum() == 0, "Free trajectories cannot have collisions"
        

    # Find range of collisions
    indices = joint_torques_df[joint_torques_df['collisions'] == 1].index

    #print(indices)

    # Find start and end indices
    _collision_range = []
    if len(indices) != 0:
        start_index = indices[0]
        for i in range(1, len(indices)):
            if indices[i] != indices[i-1] + 1:
                _collision_range.append([start_index, indices[i-1]])
                start_index = indices[i]
        
        # Add the last group
        _collision_range.append([start_index, indices[-1]])
    all_collision_ranges.append(copy.deepcopy(_collision_range))
    
    print(expt_idx,":Collision index range:", _collision_range)
```

```python
def plot_dof_torques_tiled(dof_forces_seq,  coll_range=[], y_scale=()):
    matplotlib.rc_file_defaults()
    sns.set_theme()
    
    matplotlib.rc('xtick', labelsize=14)
    matplotlib.rc('ytick', labelsize=14)
    colors = ['red', 'green', 'blue', 'orange', 'purple', 'brown']
    
    fig, axs = plt.subplots(2, 3, figsize=(10, 8))
    fig.suptitle("\nJoint Torques for each Kinova Dofs", fontsize=16)

    for i in range(2):
        for j in range(3):
            index = i * 3 + j
            sm_dof_forces_seq= dof_forces_seq[:, index]
            f_tensor_np = sm_dof_forces_seq.cpu().numpy()
            
            axs[i, j].plot(range(len(f_tensor_np)), f_tensor_np)
            axs[i, j].set_xlabel('Time')
            axs[i, j].set_ylabel('Torques')
            axs[i, j].set_title(f"Dof-{index+1}")
            if len(y_scale) !=0:
                axs[i, j].set_ylim(y_scale)
                
            if coll_range is not None:
                assert isinstance(coll_range, list) and all(isinstance(sublist, list) for sublist in coll_range), "Expecting list of lists"
                for r_idx, r in enumerate(coll_range):
                    axs[i, j].axvline(x=r[0], color=colors[r_idx], label='Collisions', alpha = 0.9)
                    axs[i, j].axvline(x=r[1], color=colors[r_idx], label='Collisions', alpha = 0.9)

    
    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.show()
```

```python
def calculate_df_freq(df):
    """
    Function to calculate time differences between consecutive rows in a DataFrame.

    Parameters:
    - df: DataFrame, the input DataFrame with a time column.

    Returns:
    - time_diff: Series, the time differences between consecutive rows.
    - frequency: float, the frequency calculated from the time differences.
    """
    # Convert time column to datetime
    df['time'] = pd.to_datetime(df['time'], unit='s')

    #df = df.sort_values(by='time')

    # Calculate the difference in seconds between consecutive rows
    df['time_diff_seconds'] = df['time'].diff().dt.total_seconds()

    # Extract time differences
    time_diff = df['time_diff_seconds']

    # Calculate frequency
    frequency = 1 / time_diff.mean()

    return time_diff, frequency

# Example usage:
# Assuming joint_state_df is your DataFrame
sample_idx = 10
joint_state_df_sample_clone = all_expt_obs_frames[sample_idx].copy()
js_time_diff, js_frequency = calculate_df_freq(joint_state_df_sample_clone)

joint_cmd_vel_df_sample_clone = all_expt_cmd_vel_frames[sample_idx].copy()
cjv_time_diff, cjv_frequency = calculate_df_freq(joint_cmd_vel_df_sample_clone)

print("Frequency of Observed Joint States for a sample :", js_frequency)
print("Frequency of Commanded Joint velocity for a sample :", cjv_frequency)

cmd_vel_slide_value = int(cjv_frequency//js_frequency)
print("Number of slides to take for cmd vel :", cmd_vel_slide_value)
```

```python
def compute_sliding_averages(df, time_col, _cmd_vel_columns, window_size=5):
    """
    Calculate sliding averages for joint1 and joint2 columns based on time column.

    Parameters:
        df (DataFrame): Input DataFrame.
        time_col (str): Name of the column containing time values.
        window_size (int): Size of the sliding window.

    Returns:
        DataFrame: DataFrame with sliding averages added as new columns.
    """
    # Set time as index
    df['idx_time'] = df[time_col]
    df.set_index('idx_time', inplace=True)

    for _cvc in _cmd_vel_columns:
        df[f'{_cvc}_last_n'] = df[_cvc].rolling(window=window_size, min_periods=1).mean()
        

for expt_idx in range(num_total_expts):
    temp_cmd_vel_frame = all_expt_cmd_vel_frames[expt_idx]  
    compute_sliding_averages(temp_cmd_vel_frame, 'time', cmd_vel_columns, window_size=cmd_vel_slide_value)
```

### Create a Single DF 

```python
def merge_df_on_closest_tme(obs_df, cmd_df):
    assert all(col in obs_df.columns for col in obs_vel_columns), "Not all column names are present in the DataFrame"
    assert all(col in cmd_df.columns for col in cmd_vel_columns), "Not all column names are present in the DataFrame"
    assert all(f"{col}_last_n" in cmd_df.columns for col in cmd_vel_columns), "Not all column names are present in the DataFrame"
    
    cmd_df.drop(columns=cmd_vel_columns, inplace=True)
    
    obs_df['obs_time'] = obs_df['time'].copy()
    cmd_df['cmd_time'] = cmd_df['time'].copy()
    obs_df.sort_values(by='time', ascending=True, inplace=True)
    cmd_df.sort_values(by='time', ascending=True, inplace=True)
    merged = pd.merge_asof(obs_df, cmd_df, left_on='time', right_on='time', direction='backward')
    count_rows_with_nan = merged.isna().any(axis=1).sum()
    assert count_rows_with_nan <= 1, f"More than one row has NaN values {count_rows_with_nan}"
    merged.dropna(inplace=True, how='any')
    return merged


all_expt_comb_frames = []
for expt_idx in range(num_total_expts):
    merged_df = merge_df_on_closest_tme(obs_df= all_expt_obs_frames[expt_idx].copy(), cmd_df= all_expt_cmd_vel_frames[expt_idx].copy())
    all_expt_comb_frames.append(merged_df)
```

## Data Pre Processing

```python
def build_features_for_expts(_expt_set):
    dof_vel_torque_final_ds = torch.empty(0, cc.total_feature_cnt).to(device)
    collision_label_final_ds = torch.empty(0, 1).to(device)
    
    for expt_idx in _expt_set:
         valid_joint_torques_df = all_expt_comb_frames[expt_idx]
         dof_torque_t  = torch.tensor(valid_joint_torques_df[obs_torque_columns].values, dtype=torch.float32).to(device)
    
         coll_t  = torch.tensor(valid_joint_torques_df[['collisions']].values, dtype=torch.float32).to(device)
    
         dof_torque_features = torch.empty(0, cc.total_torque_feature_cnt).to(device) 
         coll_labels = coll_t.detach().clone()
    
         assert dof_torque_t.dim() == 2, "invalid shape"
         assert dof_torque_t.shape[1] == cc.num_kinova_dofs, "invalid shape"
    
         slide_value = cc.store_frame_lag
         for i in range(dof_torque_t.size(0)):
            start_index = max(0, i - slide_value + 1)
            end_index = i + 1
    
            curr_dof_torque_t_w = dof_torque_t[start_index:end_index]
    
            dof_torque_features_w = cc.build_features_by_window(curr_dof_torque_t_w, n_lags=cc.store_frame_lag, device=device)
            dof_torque_features = torch.cat((dof_torque_features, dof_torque_features_w), dim=0)
    
         assert dof_torque_features.shape[0] == coll_labels.shape[0], "dataset and label shape should match"
    
         vel_feature_cols = [*obs_vel_columns, *sliding_cmd_vel_columns] # dont change this order
         dof_vel_t = torch.tensor(valid_joint_torques_df[vel_feature_cols].values, dtype=torch.float32).to(device)
         assert dof_torque_features.shape[0] == dof_vel_t.shape[0], "Mismatch in vel and torque features."
    
         dof_vel_torque_features = torch.cat((dof_vel_t, dof_torque_features), dim=1)
     
         dof_vel_torque_final_ds = torch.cat((dof_vel_torque_final_ds, dof_vel_torque_features.squeeze(0)), dim=0)
         collision_label_final_ds = torch.cat((collision_label_final_ds, coll_labels.squeeze(0)), dim=0)
    
    
    print(dof_vel_torque_final_ds.shape)
    assert dof_vel_torque_final_ds.shape[0] == collision_label_final_ds.shape[0], "dataset and label shape should match"
    return dof_vel_torque_final_ds, collision_label_final_ds
```

### Paper Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_roc_auc(y_test, pred_outputs_rf, pred_outputs_nn):
    matplotlib.rc_file_defaults()
    sns.set_theme()
    #sns.set_style("lightgrid")

    matplotlib.rc('xtick', labelsize=10) 
    matplotlib.rc('ytick', labelsize=10) 

    fpr_rf, tpr_rf, _ = roc_curve(y_test.cpu().numpy(), pred_outputs_rf.cpu().numpy())
    auc_rf = auc(fpr_rf, tpr_rf)
    
    fpr_nn, tpr_nn, _ = roc_curve(y_test.cpu().numpy(), pred_outputs_nn.cpu().numpy())
    auc_nn = auc(fpr_nn, tpr_nn)
    
    plt.figure(figsize=(8, 6))
    
    plt.plot(fpr_rf, tpr_rf, color=sns.color_palette().as_hex()[0], lw=2, label='Random Forest (AUC = %0.2f)' % auc_rf)
    plt.plot(fpr_nn, tpr_nn, color=sns.color_palette().as_hex()[1], lw=2, label='Neural Network (AUC = %0.2f)' % auc_nn)
    
    plt.plot([0, 1], [0, 1], color='black', lw=1, linestyle='--')
    
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    #plt.title('Receiver Operating Characteristic (ROC) Curve')
    plt.legend(loc="lower right")


# Assuming you have y_test, rf_pred_outputs_test, nn_pred_outputs_test as torch tensors
plot_roc_auc(y_test, rf_pred_outputs_test, nn_pred_outputs_test)
```

```python

def plot_confusion_matrices(y_test, pred_outputs_rf, pred_outputs_nn):
    matplotlib.rc_file_defaults()
    sns.set_theme()

    matplotlib.rc('xtick', labelsize=10) 
    matplotlib.rc('ytick', labelsize=10) 
    plt.figure(figsize=(8, 6))
    
    # Confusion Matrix for Random Forest
    plt.subplot(2, 1, 1)  # 2 rows, 1 column
    cm_rf = confusion_matrix(y_test.cpu().numpy(), (pred_outputs_rf.cpu().numpy() > 0.5).astype(int))
    cm_rf = cm_rf.astype('float') / cm_rf.sum(axis=1)[:, np.newaxis] # Normalize confusion matrix
    sns.heatmap(cm_rf, annot=True, cmap='Blues', fmt='.1%', cbar=False, square=True, cbar_kws={"shrink": 0.5})
    plt.title('Random Forest')
    plt.xlabel('')
    plt.ylabel('Actual')
    
    # Confusion Matrix for Neural Network
    plt.subplot(2, 1, 2)  # 2 rows, 1 column
    cm_nn = confusion_matrix(y_test.cpu().numpy(), (pred_outputs_nn.cpu().numpy() > 0.5).astype(int))
    cm_nn = cm_nn.astype('float') / cm_nn.sum(axis=1)[:, np.newaxis] # Normalize confusion matrix
    sns.heatmap(cm_nn, annot=True, cmap='Blues', fmt='.1%', cbar=False, square=True, cbar_kws={"shrink": 0.5})
    plt.title('Neural Network')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    
    plt.tight_layout()


# Assuming you have y_test, rf_pred_outputs_test, nn_pred_outputs_test as torch tensors
plot_confusion_matrices(y_test, rf_pred_outputs_test, nn_pred_outputs_test)
```

```python
def plot_dof_torques_tiled_eval(dof_forces_seq,  _pred_coll_indices=None, _true_coll_range = None , y_scale=()):
    matplotlib.rc_file_defaults()
    sns.set_theme()
    
    matplotlib.rc('xtick', labelsize=14)
    matplotlib.rc('ytick', labelsize=14)
    colors = ['red', 'green', 'blue', 'orange', 'purple', 'brown']
    
    fig, axs = plt.subplots(2, 3, figsize=(10, 8))
    fig.suptitle("\nJoint Torques for each Kinova Dofs", fontsize=16)

    for i in range(2):
        for j in range(3):
            index = i * 3 + j
            sm_dof_forces_seq= dof_forces_seq[:, index]
            f_tensor_np = sm_dof_forces_seq.cpu().numpy()
            
            axs[i, j].plot(range(len(f_tensor_np)), f_tensor_np)
            axs[i, j].set_xlabel('Time')
            axs[i, j].set_ylabel('Torques')
            axs[i, j].set_title(f"Dof-{index+1}")
            if len(y_scale) !=0:
                axs[i, j].set_ylim(y_scale)
                
            if _pred_coll_indices is not None:
                for r_idx, r in enumerate(_pred_coll_indices):
                    axs[i, j].axvline(x=r, color=colors[3], linestyle='--', label='Collisions', alpha = 0.1)

            if _true_coll_range is not None:
                assert isinstance(_true_coll_range, list) and all(isinstance(sublist, list) for sublist in _true_coll_range), "Expecting list of lists"
                for r_idx, r in enumerate(_true_coll_range):
                    axs[i, j].axvline(x=r[0], color=colors[r_idx], label='Collisions', alpha = 0.9)
                    axs[i, j].axvline(x=r[1], color=colors[r_idx], label='Collisions', alpha = 0.9)

    
    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.show()
```

```python
all_eval_accuracy = []
for curr_eval_traj in range(num_total_expts):
    curr_dof_torque_mem = torch.empty(0, cc.num_kinova_dofs).to(device)
    curr_eval_joint_torques_df = all_expt_comb_frames[curr_eval_traj]
    curr_dof_torque_t  = torch.tensor(curr_eval_joint_torques_df[obs_torque_columns].values, dtype=torch.float32).to(device)
    
    curr_obs_vel_features_t = torch.tensor(curr_eval_joint_torques_df[obs_vel_columns].values, dtype=torch.float32).to(device)
    curr_sliding_cmd_vel_features_t = torch.tensor(curr_eval_joint_torques_df[sliding_cmd_vel_columns].values, dtype=torch.float32).to(device)
    
    curr_coll_t  = torch.tensor(curr_eval_joint_torques_df[['collisions']].values, dtype=torch.float32).to(device)
    
    curr_eval_pred_collisions = []
    for _row_idx in range(curr_dof_torque_t.shape[0]):
        dof_torque_t = curr_dof_torque_t[_row_idx:_row_idx+1,:]
        curr_dof_torque_mem = torch.cat((curr_dof_torque_mem, dof_torque_t), dim=0)
        curr_dof_torque_mem = curr_dof_torque_mem[-cc.store_frame_lag:]

        curr_eval_pred, curr_eval_label = rki.real_collision_prob(loaded_model=sel_classifier, curr_dof_torque_mem=curr_dof_torque_mem, classifier_type = sel_c,
                                             obs_vel_features = curr_obs_vel_features_t[_row_idx:_row_idx+1,:],
                                             sliding_cmd_vel_features = curr_sliding_cmd_vel_features_t[_row_idx:_row_idx+1,:])
        curr_eval_pred_collisions.append(curr_eval_label)
    
    curr_eval_accuracy = accuracy_score(curr_coll_t.cpu().numpy(), torch.tensor(curr_eval_pred_collisions).unsqueeze(1).cpu().numpy())
    all_eval_accuracy.append(curr_eval_accuracy)
    print(curr_eval_traj,":",round(curr_eval_accuracy*100, 2), "%")
print("Accuracy Mean over all trajs:",round(np.mean(all_eval_accuracy)*100, 2), "%")
```
